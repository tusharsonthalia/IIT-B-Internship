```python
import pandas as pd
import numpy as np
```

```python
loc = 'D:\BDS DOCS\Year 1\Internship\Covid Data\Zonewise_Data_COVID19.csv'
```

```python
df = pd.read_csv(loc)
```

```python
df
```

```python
states = df['State'].unique()
```

```python
reds = []
greens = []
yellows = []

for state in states:
    temp = df[df['State'] == state]
    red = temp[temp['Category'] == 'Red'].__len__()
    green = temp[temp['Category'] == 'Green'].__len__()
    yellow = temp[temp['Category'] == 'Yellow'].__len__()
    reds.append(red)
    greens.append(green)
    yellows.append(yellow)
```

```python
data = {'State': states,
       'Red': reds,
       'Yellow': yellows,
       'Green': greens}
```

```python
df = pd.DataFrame(data)
```

```python
df['Total'] = df['Red'] + df['Yellow'] + df['Green']
```

```python
df['Red(%)'] = (df['Red'] * 100) / df['Total']  
df['Yellow(%)'] = (df['Yellow'] * 100) / df['Total']  
df['Green(%)'] = (df['Green'] * 100) / df['Total']  
```

```python
df.to_csv('D:\BDS DOCS\Year 1\Internship\Covid Data\Zonewise_Percentage_Data_COVID19.csv', index=False)
```
